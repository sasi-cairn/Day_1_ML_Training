### 1.1 Aritificial Intelligence

"Aritifical Intelligence involves machines that can perform tasks that are characteristic of human intelligence". Aritifical Intelligence are powered by algorithms, using techniques such as machine learning, deep learning and rules. 

Aritifical Intelligence is further divided into three branches based on thier capacity to mimic human characteristics

<b>1.1.a Artificial Narrow Intelligence</b>

It is also referred as Weak AI (or) Narrow AI. This is the stage where we have succefully realized to this date. Narrow AI is goal-oriented, designed to perform singular tasks

<b>i.e</b> Facial recognition, Speech Recognition, Self driving cars

The algorithms operate under a narrow set of constraints and limitations, which is why it is reffered as Weak AI. It doesn't replicate human intelligence, it merely simulates human behaviour based on narrow range of parameters.

<b>1.1.b Artificial General Intelligence</b>

It is also referred as Strong AI. It is considered to mimic human intelligence, with it's ability to learn and apply it's intelligence to solve any problem. It can act in a way that is indistinguishable from that of a human in any given situation.

<b>1.1.c Artificial Super Intelligence</b>

This is hypothetical AI, where it doesn't mimic the human behaviour, it will become self-aware and surpass the capcity of human intelligence and ability.

### 1.4 Data Science

Data Science is a field which deals with all kind of data (Structured, Semi-Structured and Unstructured Data) and all aspects of data (Data Visualization, Statistics, Social Sciences, Data Engineering etc) to uncover useful insights. 

### 1.5 Data Analytics

Data Analytics is one of the discipline of Data Science. It is dividied into 5 major sub-divsions, which has different objectives ranging from identifying the problem, uncovering the hidden trends to overall improvement in the process of the business

1. Descriptive Analytics - It gives insights on day-to-day business operations. This helps us to identify deviations, trends.


2. Diagnostics Analytics - It helps us to identify the root cause of the deviation (or) anamoly and explain the certian behavioural patterns.


3. Predictive Analytics - It helps to predict the events based on the historical trends observed.


4. Prescriptive Analytics - It helps to provide recommendations to avoid events from happening based on it's learning from the past.


5. Cognitive Analytics - It helps to perform (or) simulate human like behaviour in acheiveing certian tasks.

#### 1.7.4 Training Set,  Validation Set and Test Set

   <b>Training Set</b> - The actual dataset that we use to train the model
   
   <b>Validation Set</b> - The sample of data used to provide an unbiased evaluation of a model fit on the training dataset while tuning model hyperparameters.

   <b>Test Set</b> - The sample of data used to provide an unbiased evaluation of a final model fit on the training dataset.

#### 1.7.5 Dependent Variable & Independent Variable

<b>Independent Variables</b> - It is the cause. It's value is independent of other variables.

<b>Dependent Variables</b> - It is the effect. Its value depends on changes in the independent variable.

#### 1.7.6 Types of Data

1. <b>Qualitative Data Type: </b> It describes the object under consideration using a finite set of discrete classes. 

    1. Nominal Data - These are the set of values that dont possess a natural ordering.
    2. Ordinal Data - These types of values have a natural ordering while maintaining their class of values.

2. <b>Quantitative Data Type: </b> It tries to quantify things and it does by considering numerical values

    1. Discrete Data - The numerical values which fall under are integers or whole numbers are placed under this category. 
    2. Continous Data - The fractional numbers are considered as continuous values.
