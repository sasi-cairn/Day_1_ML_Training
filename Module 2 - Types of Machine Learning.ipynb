### 2.1 Supervised Learning

The training set that you feed to the algorithms includes the desired solutions, called "labels". These labels will act as a judge and supervise the algorithm in finding the right answer.

Y = f(X)

Y = Target variable

f(X) = a.x1 + b.x2 + c.x3 (Independent Variables) = Mapping Function

The data points are present in the n-dimensional space. The fucntion is trying to learn the fitted line (or) decision boundries in the n-dimesional space, where labels acts as a judge (or) supervise to find the best fitted line (or) decision boundry by comparing the predited value (Y') and actual value (Y). 

The supervised learning can be further classified into two tasks based on the nature of the label.

1. Regression Task - Target Variable / Label is <b>Numeric</b> in nature <b>(i.e. 12, 16, 56.0)</b>

### 2.2 Unsupervised Learning

The absence of desired output in the dataset makes the problem to fall under the category of Unsupervised learning because there is no element to guide the algorithm to find the right answer. Here, the algorithm is tasked to find the answers by automatically find structure in the data by extrating useful features and analyzing it's structure.

The Unsupervised algorithms can be further classified into based on the task.

1. Clustering

   Clusering is technique to identify and group the unlabelled data labels. Datapoints exhibting similar properties are grouped in sinlge group.
   
   <b>Example: </b> Cluster analysis of 3D seismic data
   
   In order to recognize and classify reservoir characteristics - geoscientists analyze the variability in seismic reflections. This activity demands a professional able to identify subtle waveform characteristics and tools to measuring these characteristics are given by seismic attributes. So these attributes are important in reservoir characterization and the correct use of them depends on the experience of a geoscientist, that is, his understanding and interpretation of their behavior. 
   
   Clustering algorithm is applied on the sesimic attributes to idetify distint groups which are having similar attributes and each group can be characterized based on the similar properties it posses.
   
   PBM Index is the metric used to identify the best number of groups.

4. Dimensionality Reduction


A dataset contains a huge number of input features in various cases, which makes the predictive modeling task more complicated. Because it is very difficult to visualize or make predictions for the training dataset with a high number of features, for such cases, dimensionality reduction techniques are required to use.
    
Dimensionality reduction technique can be defined as, "It is a way of converting the higher dimensions dataset into lesser dimensions dataset ensuring that it provides similar information."
    
    Most common Dimesionality Reduction Technique

    1. PCA - Prinicipal Components Analysis
    
     It is a method of reducing the dimensionality of a data set by transforming it into a new coordinate system such that the greatest variance in the data is explained by the first coordinate and the second greatest variance is explained by the second coordinate, and so on

### 2.4 Reinforcement Learning

The learing is based on the reward (or) penalty. The algorithm acts as a virtual agents in the known environment where these agents choose the possible options of action. The agents selects the best action from all the possibilities present in that envionmental state and based on that selection, receives reward/risks. The algorithm keep an eye on maximzing the reward, reducing the risk and eventually learning. 

The user actions are recorded and stored in the database. The AI agent learns from this recorded data in batch mode. When the AI agent has learned enough from users actions to approximate the recommendation at high accuracy
